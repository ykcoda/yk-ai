### Using tools

```python
import httpx
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from langchain.tools import tool
load_dotenv()


@tool
def get_weather(city: str):
    """Gets the weather upon city provided by the user"""
    response = httpx.get(f"https://wttr.in/{city}?format=j1", timeout=20)
    return response.json()

agent = create_agent(model="gpt-5-mini", tools=[get_weather])

response = agent.invoke({"messages":[HumanMessage(content="What is the weather like in kumasi")]})
print(response['messages'])

```
