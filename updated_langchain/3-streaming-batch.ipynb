### Streaming

```python
import os
from dotenv import load_dotenv
from langchain.agents import create_agent 
from langchain_openai import ChatOpenAI
load_dotenv()
```

```python
model = ChatOpenAI()
for chunk in model.stream("Write a 200 words paragraph on artificail interlligent"):
    print(chunk.text, end="", flush=True)
```

```python
model = ChatOpenAI()
for chunk in model.stream("Why do Ghanains like fufu"):
    print(chunk.text, end="", flush=True)
```

### Batch

```python
responses = model.batch({
    "Why do africans love football?",
    "How do airplanes fly?",
    "what is AI Engineering?"
})

for response in responses:
    print(response)
```
