```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.agents import create_agent


def get_weather(city:str)-> str:
    """Get weather for a city"""
    return "Weather in the {city} is sunny."
    
agent = create_agent(model="gpt-5-mini", tools=[get_weather], system_prompt="You are a helpful assistant.")
agent
```

```python
response = agent.invoke({"messages":[{"role":"user", "content":"What is the weather in New york"}]})
```

```python
response["messages"][-1].content
```

```python
response = agent.invoke({"messages":"What is the weather in Accra"})
response["messages"][-1].content
```
